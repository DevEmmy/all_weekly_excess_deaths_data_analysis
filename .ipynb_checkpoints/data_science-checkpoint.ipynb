# Weekly Deaths in different Countries
There are a lot of deaths occuring almost everyday in each country which is caused by accidents, poverty, diseases (or virus), etc. These deaths were taken down into data from every part of the country so as to understand what causes the deaths and the amount of deaths per week. Now this data is going to go through a process called "Data Wrangling" to clean the data up for its analysis.For this process, I used a python library called "Pandas" which is used majorly in "Data Science" to access files and also perform different operations with it. So i start by importing as 'pd' as an alias to save the stress of typing it in full.

```python
import pandas as pd
```

Then the next thing is to open the required file using the built in function in pandas which is "read_csv()" to open the file - specifically csv file -to begin the wrangling operation with it, then print the first 5 rows.

```python
file = pd.read_csv('weekly_deaths.csv')
file.head()
```

The next step is to check if we have a Null value in any of the column by using the "isnull()" function.

```python
file.isnull().sum()
```

So since there is no Null value in the data set there is no need of calling the "dropna()" method which is used to drop the row which includes a Null value. If you observe the data, the region and country has the same values, so i'd just remove the 'region' column from the whole data.

```python
file.drop('region',axis=1, inplace=True)
file.head()
```

Now to check the whole information about the data set, i use the 'info()' function.

```python
file.info()
```

```python
file.describe()
```
